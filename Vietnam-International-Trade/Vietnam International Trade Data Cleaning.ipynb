## Load Dataset

HS6 REV. 2007 (2008 - 2017) https://atlas.media.mit.edu/en/resources/data/

```python
data = pd.read_csv("year_origin_hs07_4.tsv",  delimiter="\t") #Product Trade by Year and Country (4 digit depth)
data.head()
```

```python
data.tail()
```

```python
indo = data.loc[data["origin"] == "idn"]
viet = data.loc[data["origin"] == "vnm"]
print (indo.shape , viet.shape)
indo.info()
```

```python
indo.tail()
```

```python
indo.head()
```

```python
#Grouping World Trade by year
def siapa(frame, tahun):
    x = frame.loc[frame["year"] == tahun]
    x["hs07"] = (x["hs07"] / 100).round()
    x = x.groupby("hs07").sum()
    x["year"] = tahun
    x = x.reset_index()
    x["export_val_perc"] = x["export_val"] / x["export_val"].sum()
    x["import_val_perc"] = x["import_val"] / x["import_val"].sum()
    return x
```

```python
i2008 = siapa(indo, 2008)
i2009 = siapa(indo, 2009)
i2010 = siapa(indo, 2010)
i2011 = siapa(indo, 2011)
i2012 = siapa(indo, 2012)
i2013 = siapa(indo, 2013)
i2014 = siapa(indo, 2014)
i2015 = siapa(indo, 2015)
i2016 = siapa(indo, 2016)
i2017 = siapa(indo, 2017)
v2008 = siapa(viet, 2008)
v2009 = siapa(viet, 2009)
v2010 = siapa(viet, 2010)
v2011 = siapa(viet, 2011)
v2012 = siapa(viet, 2012)
v2013 = siapa(viet, 2013)
v2014 = siapa(viet, 2014)
v2015 = siapa(viet, 2015)
v2016 = siapa(viet, 2016)
v2017 = siapa(viet, 2017)
w2008 = siapa(data, 2008)
w2009 = siapa(data, 2009)
w2010 = siapa(data, 2010)
w2011 = siapa(data, 2011)
w2012 = siapa(data, 2012)
w2013 = siapa(data, 2013)
w2014 = siapa(data, 2014)
w2015 = siapa(data, 2015)
w2016 = siapa(data, 2016)
w2017 = siapa(data, 2017)
```

```python
indonesia = pd.concat([i2008, i2009, i2010, i2011, i2012, i2013, i2014, i2015, i2016, i2017], ignore_index=True)
indonesia["country"] = "indonesia"
vietnam = pd.concat([v2008, v2009, v2010, v2011, v2012, v2013, v2014, v2015, v2016, v2017], ignore_index=True)
vietnam["country"] = "vietnam"
bilateral = pd.concat([indonesia, vietnam], ignore_index=True)
bilateral.head()
```

```python
bilateral.to_csv("bilateral.csv")
```

```python
world = pd.concat([w2008, w2009, w2010, w2011, w2012, w2013, w2014, w2015, w2016, w2017], ignore_index=True)
world.head()
```
